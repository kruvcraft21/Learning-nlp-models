3) Напишите функцию Python, которая находит шаблон вида :, за которой следует один или несколько )

Нужно найти **шаблон вида `:` за которым идёт одна или более закрывающих скобок `)`** — это классический смайлик `:)`, `:))`, `:)))` и т.п.

Для этого используем регулярное выражение:

```
r":\)+"
```

- `:` — ищем буквально двоеточие.
- `\)` — экранируем скобку, так как `)` в регэкспе имеет особый смысл.
- `+` — означает "один или больше".

````python
def find_smileys(text: str) -> list[str]:
    """
    Ищет смайлики вида :), :)), :))) и т.п.
    Возвращает список найденных совпадений.
    """
    pattern = r":\)+"
    return re.findall(pattern, text)


# Примеры, где должно находить
print(find_smileys("Привет :)"))                # [':)']
print(find_smileys("Вот так :)) класс"))        # [':))']
print(find_smileys("Ха-ха :))) :D"))            # [':)))']
print(find_smileys('Несколько смайлов :)))))))))), был первый, а сейчас второй :)'))

# Примеры, где находить не должно
print(find_smileys("Грустный смайл :("))        # []
print(find_smileys("Текст без смайлов"))        # []
print(find_smileys("Эмодзи 😊 вместо текста"))  # []
````

9) Напишите функцию Python, которая находит шаблон вида l, за которым следует что-либо, заканчивающееся на e

Нужно найти **шаблон вида `l`, за которым следует что-либо, заканчивающееся на `e`**.

Для этого используем регулярное выражение:

`r"l.*?e"`

- `l` — ищем букву `l` как начало шаблона.
- `.*?` — означает "любые символы (кроме перевода строки), 0 или более раз, как можно меньше". Мы используем ленивый квантификатор, чтобы не захватывать лишнее.
- `e` — ищем букву `e` в конце шаблона.

````python
import re

def find_l_to_e(text: str) -> list[str]:
    """
    Ищет все подстроки в тексте, которые начинаются с 'l',
    за которыми следует что угодно, и заканчиваются на 'e'.
    
    :param text: Строка для поиска
    :return: Список найденных подстрок
    """
    # Регулярное выражение:
    # l       - ищем букву 'l'
    # .*?     - любое количество любых символов (нежадный поиск)
    # e       - заканчивается на 'e'
    pattern = r"l.*?e"

    # findall возвращает все найденные совпадения
    matches = re.findall(pattern, text)
    return matches

# Примеры корректной работы
print(find_l_to_e("like"))        # ['like'] — l ... e
print(find_l_to_e("apple"))       # ['le'] — l ... e
print(find_l_to_e("lovely"))      # ['love'] — l ... e
print(find_l_to_e("little"))      # ['little'] 

# Примеры, где совпадения нет
print(find_l_to_e("lion"))        # [] — нет окончания на 'e'
print(find_l_to_e("test"))        # [] — нет буквы 'l'
print(find_l_to_e("laptop"))      # [] — есть 'l', но нет 'e' после него
````

12) Напишите функцию Python, которая находит слово, содержащее q

Нужно найти **слово, содержащее букву `q`**.

Для этого используем регулярное выражение:

`r"\b\w*q\w*\b"`

- `\b` — граница слова, чтобы не захватывать часть слова;
- `\w*` — любое количество букв/цифр до и после буквы `q`;
- `q` — сама буква, которую ищем.

````python
def find_word_with_q(text) -> list[str]:
    """
    Находит все слова, содержащие букву 'q' в переданном тексте.
    
    :param text: Строка для поиска
    :return: Список слов с буквой 'q'
    """
    # Используем re.findall с шаблоном \b\w*q\w*\b
    # \b — границы слова, \w* — любая последовательность букв/цифр, q — ищем букву q
    pattern = r'\b\w*q\w*\b'
    return re.findall(pattern, text)


# Примеры, когда регулярка работает (слово содержит 'q')
print(find_word_with_q("The quick brown fox jumps over the lazy dog"))  # ['quick'] — слово "quick" содержит 'q'
print(find_word_with_q("I acquired a unique qualification"))            # ['acquired', 'unique', 'qualification'] — несколько слов с 'q'
print(find_word_with_q("No question here"))                             # ['question'] — слово "question" содержит 'q'


# Примеры, когда регулярка не работает (слово не содержит 'q')
print(find_word_with_q("Hello world"))                                  # [] — нет слов с 'q'
print(find_word_with_q("Python programming"))                           # [] — нет слов с 'q'
print(find_word_with_q("Data science is fun"))                          # [] — нет слов с 'q'
````

15) Напишите функцию Python которая ищет шаблон вида начинается с + определенного числа (пусть это будет 7)

Нужно найти строки, которые **начинаются с `+7`**. 

Для этого используем регулярное выражение:

`\+7\d+`

- `\+` экранирует знак плюса (иначе он будет иметь специальное значение в regex), `7` — просто цифра 7.
- `\d+` ищет цифры после `+7`, подумал что речь идет про номер, но номер может быть обрезан.

````python

def find_plus7(text: str) -> list:
    """Ищет в тексте все вхождения, начинающиеся с +7 и продолжающиеся цифрами."""
    pattern = r"\+7\d+"
    return re.findall(pattern, text)

# Примеры использования:

print(find_plus7("+71234567890"))   # ['+71234567890'] — начинается с +7 и дальше только цифры, находит номер
print(find_plus7("Телефон +7123"))  # ['+7123'] — начинается с +7 и цифры, работает
print(find_plus7("+7123abc"))       # ['+7123'] — находит только часть с цифрами после +7

print(find_plus7("7+1234567890"))   # [] — не начинается с +7, а находит 7+, не подходит
print(find_plus7("+81234567890"))   # [] — начинается с +8, не с +7, не ищет
print(find_plus7("напиши +7 и что-то"))  # [] — после +7 нет цифр, не будет найдено
print(find_plus7("номер +7"))       # [] — +7 без цифр после, не подходит
````

16) Напишите функцию Python для удаления ведущих нулей из IP-адреса.

например 255.01.092.132 -> 255.1.92.132

Нужно найти **все ведущие нули в каждом октете IP-адреса**, чтобы преобразовать, например, `255.01.092.132` в `255.1.92.132`.

Для этого используем регулярное выражение:

- Ищем каждую последовательность цифр, которая может начинаться с нулей, но не должна превращать одиночный `0` в пустую строку.
- Используем `\b0+(\d+)` – `\b` гарантирует границу числа, `0+` — один или несколько ведущих нулей, а `(\d+)` — оставшиеся цифры, которые нужно оставить.
- Функция `re.sub` заменяет все найденные случаи на `\1` (оставшиеся цифры без ведущих нулей).

````python
def find_strip_leading_zeros(ip_address: str) -> str:
    """
    Убирает ведущие нули из каждого октета IP-адреса.
    
    :param ip_address: IP-адрес в виде строки, возможно с ведущими нулями.
    :return: IP-адрес без ведущих нулей.
    """
    # Используем регулярное выражение для поиска ведущих нулей перед цифрами
    # \b0+(\d+) - граница числа, один или несколько нулей, захватываем оставшиеся цифры
    return re.sub(r'\b0+(\d+)', r'\1', ip_address)


# Примеры работы функции (работает как надо)
print(find_strip_leading_zeros("255.01.092.132"))   # 255.1.92.132
print(find_strip_leading_zeros("010.002.003.004"))  # 10.2.3.4
print(find_strip_leading_zeros("192.168.001.001"))  # 192.168.1.1
print(find_strip_leading_zeros("0.0.0.0"))          # 0.0.0.0

# Примеры, когда регулярка не сработает (не надо удалять)
print(find_strip_leading_zeros("192.168.0.1"))      # 192.168.0.1, всё ок
print(find_strip_leading_zeros("0.255.0.001"))      # 0.255.0.1, корректно
print(find_strip_leading_zeros("127.0.0.1"))        # 127.0.0.1, корректно
````

20) Напишите функцию Python для поиска строки литералов в строке, а также найдите место в исходной строке, где встречается шаблон

Пример текста: 'The quick brown fox jumps over the lazy dog.'

Искомые слова: 'lazy'

Нужно найти **строку литералов в строке и её позицию**.

Для этого используем регулярное выражение:

`rf'\b{re.escape(word)}\b'`

- `\b` — это граница слова. Она нужна, чтобы регулярка не находила части других слов, например, `lazy` или `dog`.

````python
def find_literal_word(text, word):
    """
    Ищет слово в тексте и возвращает его позицию.

    :param text: Исходная строка
    :param word: Слово для поиска
    :return: Список кортежей (найденное слово, start, end)
    """
    # Формируем шаблон с границами слова
    pattern = rf'\b{re.escape(word)}\b'

    # Ищем все вхождения
    matches = [(m.group(), m.start(), m.end()) for m in re.finditer(pattern, text)]

    return matches


# Примеры использования
text = 'The quick brown fox jumps over the lazy dog.'

print(find_literal_word(text, 'lazy'))                  # [('lazy', 35, 39)]

# Примеры, когда регулярка работает
print(find_literal_word('A lazy cat sleeps.', 'lazy'))  # [('lazy', 2, 6)]
print(find_literal_word('lazy', 'lazy'))                # [('lazy', 0, 4)]
print(find_literal_word('It is a lazy-day.', 'lazy'))   # [('lazy', 8, 12)]
text3 = 'lazy dog and another lazy dog'
print(find_literal_word(text3, 'lazy'))                 # [('lazy', 0, 4), ('lazy', 21, 25)]
text4 = 'A cat, a dog, and another dog.'
print(find_literal_word(text4, 'dog'))                  # [('dog', 9, 12), ('dog', 26, 29)]

# Примеры, когда регулярка не работает (не должна находить)
print(find_literal_word('blazy dog', 'lazy'))           # [] — "lazy" является частью другого слова
print(find_literal_word('lazying around', 'lazy'))      # [] — "lazy" входит в слово "lazying"
print(find_literal_word('crazy', 'lazy'))               # [] — "lazy" нет в слове
````

25) Напишите функцию Python для преобразования даты формата гггг-мм-дд в формат дд-мм-гггг

Нужно найти даты в формате гггг-мм-дд и преобразовать их в формат дд-мм-гггг.

Для этого используем регулярное выражение: 

`r'(\d{4})-(\d{2})-(\d{2})'`

- `(\d{4})` - захватывающая группа для 4 цифр (год)
- `-` - дефис-разделитель
- `(\d{2})` - захватывающая группа для 2 цифр (месяц)
- `(\d{2})` - захватывающая группа для 2 цифр (день)

````python
def find_convert_date(date_str: str) -> str:
    """
    Преобразует дату из формата гггг-мм-дд в дд-мм-гггг.

    :param date_str: Входная строка с датой.
    :return: Дата в формате дд-мм-гггг, если вход соответствует шаблону.
         Иначе возвращает исходную строку без изменений.
    """
    # Регулярка ищет: 4 цифры (год), дефис, 2 цифры (месяц), дефис, 2 цифры (день)
    pattern = r'^(\d{4})-(\d{2})-(\d{2})$'
    
    # Замена с перестановкой: день, месяц, год
    return re.sub(pattern, r'\3-\2-\1', date_str)

# Примеры, когда регулярка работает (соответствует формату гггг-мм-дд)
print(find_convert_date("2025-09-23"))  # 23-09-2025
print(find_convert_date("1999-12-31"))  # 31-12-1999
print(find_convert_date("0001-01-01"))  # 01-01-0001

# Примеры, когда регулярка не сработает (не соответствует формату)
print(find_convert_date("23-09-2025"))  # 23-09-2025, не меняется
print(find_convert_date("2025/09/23"))  # 2025/09/23, не меняется
print(find_convert_date("2025-9-23"))   # 2025-9-23, не меняется
print(find_convert_date("Sept-23-2025"))# Sept-23-2025, не меняется
````

26) Напишите функцию Python для удаления кавычек " из строки

Нужно найти: все кавычки типа " в строке и удалить их

Для этого используем регулярное выражение:

`r'"'`

- `"` - просто ищем символ двойной кавычки

````python
def find_remove_quotes(text: str) -> str:
    """
    Удаляет все двойные кавычки из строки.
    
    :param text: исходная строка
    :return: строка без двойных кавычек
    """
    # Заменяем все " на пустую строку
    return re.sub(r'"', '', text)


# Примеры, где регулярка работает как нужно
print(find_remove_quotes('Это "пример" текста'))    # Это пример текста
print(find_remove_quotes('"Начало и конец"'))       # Начало и конец
print(find_remove_quotes('Тут "много" "кавычек"'))  # Тут много кавычек

# Примеры, где регулярка не должна удалять ничего (нет кавычек)
print(find_remove_quotes('Нет кавычек здесь'))      # Нет кавычек здесь
print(find_remove_quotes('Одинарные кавычки \'\' не трогаем'))  # Одинарные кавычки '' не трогаем
print(find_remove_quotes('1234567890'))             # 1234567890
````
